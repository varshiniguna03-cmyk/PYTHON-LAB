```python
# Employee Salary Management Case Study
# Employee data
employees = {
    "Ravi": 45000,
    "Sneha": 52000,
    "Amit": 60000,
    "Priya": 48000
}

print("Original Employee Salaries:")
for name, salary in employees.items():
    print(name, ":", salary)

# Increase salary by 10%
print("\nSalaries after 10% increment:")
for name in employees:
    employees[name] = employees[name] + (employees[name] * 10 / 100)
    print(name, ":", employees[name])

# Find employees earning above 50000
print("\nEmployees earning above 50000:")
for name, salary in employees.items():
    if salary > 50000:
        print(name, ":", salary)

# Calculate total salary expenditure
total_salary = sum(employees.values())

print("\nTotal Salary Expenditure:", total_salary)
```
