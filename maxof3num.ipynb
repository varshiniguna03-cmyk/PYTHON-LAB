```python
# find max of 3 numbers using max() and without using max function 
def maximum(a,b,c):
    return max(a,b,c)

x=maximum(5,4,9)
print(x)
```

```python
#find max of 3 numbers using max() and with using max function 
def maximum(a,b,c):
    if a>b and a>c:
        return a
    elif b>c:
        return b
    else:
        return c

x=maximum(5,4,9)
print(x)
```
