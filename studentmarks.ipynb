```python
# a college wants to process student marks and identify toppers, failed students , and class average using functional programming
from functools import reduce 
marks = [78,45,90,32,88,67,29]
# students passed
passed = list(filter(lambda x: x>=40,marks))
#add grace marks
grace_marks = list(map(lambda x: x + 5,marks))
#total marks
total = reduce(lambda x,y: x+y, marks)
average = total / len(marks)
print("passed student marks:",passed)
print("grace marks:",grace_marks)
print("total marks:",total)
print("average marks:",average)
```
